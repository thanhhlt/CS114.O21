```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/annonimized.csv')
df.head()
```

```python
# Xoá feature không cần thiết
df = df.drop(['created_at', 'updated_at', 'judgement'], axis=1)
# Đổi tên cột
df = df.rename(columns={'concat(\'it001\',`assignment_id`)': 'assignment_id',
                        'concat(\'it001\',`problem_id`)': 'problem_id',
                        'concat(\'it001\', username)': 'username',
                        'concat(\'it001\',`language_id`)': 'class'})
df.head()
```

```python
"""DataFrame lưu dữ liệu đã xử lý từ dữ liệu ban đầu
DataFrame này lưu thông tin về 10 Assign sinh viên làm và điểm trung bình của 10 Assign đó
Mỗi Assign lưu 6 thông tin:
    + Số lần submit trung bình mỗi problem.
    + Số lần submit với status là SCORE trung bình mỗi problem.
    + Số lần submit với status là Compilation Error trung bình mỗi problem.
    + Có nộp trễ hay không.
    + Điểm trung bình mỗi problem.
    + Điểm Assign."""
columns = []
columns.append('username')
for i in range(10):
    for j in range(6):
        columns.append(str(i + 1) + '_' + str(j + 1))
columns.append('avg_score')

df_final = pd.DataFrame(columns=columns)
df_final
```

```python
# Xử lý dữ liệu ban đầu
# Duyệt qua từng sinh viên
dfs_student = df.groupby(['username'])
for id_student, data_student in dfs_student:
    # Mảng lưu thông tin các Assign
    list_assign = []

    # Duyệt qua từng Assignment
    dfs_assignment = data_student.groupby(['assignment_id'])
    for id_assign, data_assign in dfs_assignment:
        # Bỏ qua Assign gây nhiễu
        # if id_assign[0] in noise_assign:
        #     continue

        assign = [] # Mảng lưu thông tin mỗi Assign
        probs_Nsubmit = []
        probs_score = []
        probs_score_mean = []

        # Duyệt qua từng Problem
        dfs_problem = data_assign.groupby('problem_id')
        for id_prob, data_prob in dfs_problem:
            probs_Nsubmit.append(len(data_prob))
            probs_score.append(data_prob['pre_score'].max())
            probs_score_mean.append(data_prob['pre_score'].mean())

        # Đưa dữ liệu vào mảng lưu thông tin Assign
        assign.append(round(np.asarray(probs_Nsubmit).mean(), 3))
        assign.append(round(data_assign['status'].value_counts().get('SCORE', 0)/len(probs_Nsubmit), 3))
        assign.append(round(data_assign['status'].value_counts().get('Compilation Error', 0)/len(probs_Nsubmit), 3))
        assign.append(1 if (data_assign['coefficient'] < 100).any() else 0)
        assign.append(round(np.asarray(probs_score_mean).mean(), 3))
        assign.append(round(sum(probs_score)/dict_Nprob_assign[id_assign[0]], 3))

        # Trường hợp làm nhiều hơn 10 Assign, sẽ chọn Assign có điểm cao nhất
        if len(list_assign) < 10:
            list_assign.append(assign)
        else:
            # Tìm Assign đang có điểm thấp nhất
            min = float('inf')
            id_min = 0
            for i in range(len(list_assign)):
                if list_assign[i][-1] < min:
                    min = list_assign[i][-1]
                    id_min = i
            # Nếu Assign cần thêm có điểm cao hơn thì thay thế
            if assign[-1] > list_assign[id_min][-1]:
                list_assign[id_min] = assign

    # Lưu thông tin vào DataFrame
    row = [id_student[0]]
    avg_score = 0
    for i in range(len(list_assign)):
        avg_score += list_assign[i][-1]
        row.extend(list_assign[i])

    avg_score = round(avg_score/10, 3)
    row += [0] * (61 - len(row))
    row.append(avg_score)

    df_final.loc[len(df_final)] = row
```
